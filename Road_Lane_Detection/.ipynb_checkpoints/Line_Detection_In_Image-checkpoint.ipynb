## Gaussian Blur

```python
# Gaussian Blur Algo is used to blur the image or clean the image 
# Bacause our computer will work Fastly on the clean or blur image where it doen't have much
# pixels to work on
# typically to reduce image noise and reduce detail. The visual effect of this blurring 
# technique is a smooth blur resembling that of viewing the image through a translucent screen


#In image processing, a kernel, convolution matrix, or mask is a small matrix. It is used for 
#blurring, sharpening, embossing, edge detection, and more. This is accomplished by doing a 
#convolution between a kernel and an image.
```

```python
# Apply Gaussiab Blur Algo-->   5X5 is a Gaussian Filter
img_blur=cv.GaussianBlur(img_gray,(5,5),0)   # (5,5) is a Gaussian Blur matrix   1-3X3 2-5X5 
plt.imshow(img_blur,cmap="gray")             # 0 show we don't want blur in any direction x,y
```

## Canny Edge Detection

```python
#Theory- It is a multi-stage algorithm & we will go through
#        each Stages. 
#     Stages --> 1- Noise Reduction
#                2- Finding Intensity gradient of the image-->
#                   Gradient direction is always perpendicular to edges. It is rounded to one of four angles representing 
#                   vertical, horizontal and two diagonal directions.   
#                3- Non-maximum Suppression -->
#                   After getting gradient magnitude and direction, a full scan of image is done to remove any unwanted pixels 
#                   which may not constitute the edge. For this, at every pixel, pixel is checked if it is a local maximum in 
#                   its neighborhood in the direction of gradient. Check the image below:
#                   Point A is on the edge ( in vertical direction). Gradient direction is normal to the edge. Point B and C are
#                   in gradient directions. So point A is checked with point B and C to see if it forms a local maximum. 
#                   If so, it is considered for next stage, otherwise, it is suppressed ( put to zero).
#                   In short, the result you get is a binary image with "thin edges".
#                4- Hysteresis Thresholding -->
#                   This stage decides which are all edges are really edges and which are not. For this, we need two threshold 
#                   values, minVal and maxVal. Any edges with intensity gradient more than maxVal are sure to be edges and those
#                   below minVal are sure to be non-edges, so discarded. 
#                   Those who lie between these two thresholds are classified edges or non-edges based on their connectivity. 
#                   If they are connected to "sure-edge" pixels, they are considered to be part of edges. Otherwise, they are 
#                   also discarded. This stage also removes small pixels noises on the assumption that edges are long lines.
#                                                                                                                            
```

```python
# Apply Canny Edge Detection Algo-->
img_canny=cv.Canny(img_blur,120,240)  # 120-240 shows the color intensity so this img_canny will detect the images which color 
plt.imshow(img_canny,cmap="gray")     # lies in 120 - 240 range & above 240 because these 2 values are threshold values
```

## Masking the require area of img_canny

```python
# It is a fxn which returns the coordinates of the requied region
def get_vertices(img_canny):
    rows, cols = img_canny.shape[:2]
    bottom_left = [cols*0.1 , rows*0.95]
    top_left = [cols*0.45,rows*0.6]
    bottom_right = [cols*0.9,rows*0.95]
    top_right = [cols*0.55,rows*0.6]
    
    ver = np.array([[bottom_left,top_left,top_right,bottom_right]],dtype=np.int32)
    return ver
```

```python
# Creating a mask of size img_canny
mask=np.zeros_like(img_canny)
# Get the coordinates of the region where the we want to detect  the lines in given image 
points=get_vertices(img)   # get_vertices(img) is a user defined fxn which returns the coordinates of the required region
cv.fillPoly(mask,points,(255,255,255)) # here we color the region which is under the points with white color
com=cv.bitwise_and(img_canny,mask)  # It will execute the and operation b/w the bits of the 'img_canny' & 'mask' till the bits 
plt.imshow(com,cmap='gray')         # of the mask where 1 are present & rest bits are zero
```

## lines Detection in Video or in Live Car

```python
import matplotlib.pyplot as plt
import cv2 as cv
from PIL import Image
import numpy as np
from pylab import rcParams
rcParams['figure.figsize']=(10,10)
```

```python
cam=cv.VideoCapture("solidWhiteRight.mp4")
```

```python
# Gives the coordinates of vertices for masking
def get_vertices(img_canny):
    rows, cols = img_canny.shape[:2]
    bottom_left = [cols*0.1 , rows*0.95]
    top_left = [cols*0.45,rows*0.6]
    bottom_right = [cols*0.9,rows*0.95]
    top_right = [cols*0.55,rows*0.6]
    
    ver = np.array([[bottom_left,top_left,top_right,bottom_right]],dtype=np.int32)
    return ver
```

```python
# Loop which is used to display the detected line
while True:
    _,frame=cam.read()
    
    frame_rgb =cv.cvtColor(frame,cv.COLOR_BGR2RGB)
    
    frame_gray =cv.cvtColor(frame_rgb,cv.COLOR_RGB2GRAY)
    
    # Apply Gaussian Blur Algo
    frame_blur=cv.GaussianBlur(frame_gray,(5,5),0)   # (5,5) is a matrix    0 show we dont need blur in any direction x,y
    #plt.imshow(frame_blur,cmap="gray")
    
    # Apply Canny Edge Detection Algo
    frame_canny=cv.Canny(frame_blur,120,240)  # 120-240 shows the color intensity so this img_canny will detect the images which color 
    #plt.imshow(img_canny,cmap="gray")         #lies in 120 - 240 range

    # Apply Masking 
    mask=np.zeros_like(frame_canny)
    points=get_vertices(frame)
    cv.fillPoly(mask,points,(255,255,255))
    com=cv.bitwise_and(frame_canny,mask)
    #plt.imshow(com,cmap='gray')
    
    # Appliying Hough Lines Algo
    lines=cv.HoughLinesP(com,      # HoughLinesP & HoughLines are diff functions             HoughLinesP is used to detect 
                         rho=6,                                                                   # the lines in imgae
                         theta=np.pi/60,
                         threshold=140,   # Shows the lines which are having color value greater than 140  
                         lines=np.array([]),
                         minLineLength=40,
                         maxLineGap=5
                        )
    
    lines.shape # shape is of 3 lists
    
    # SO to remove the complexity we reshape the lines list
    lines=lines.reshape(-1,4)
    lines.shape
    
    dg=frame_rgb.copy()
    for l in lines:
        x1,y1,x2,y2=l
        cv.line(dg,(x1,y1),(x2,y2),(255,0,0),5)
    cv.imshow("Line_Detection",dg)
    
     #To close the window
    key=cv.waitKey(5)
    if key & 0xff == ord('q'):
        cv.destroyAllWindows()
        break
#Release the camera
cam.release()  
```
